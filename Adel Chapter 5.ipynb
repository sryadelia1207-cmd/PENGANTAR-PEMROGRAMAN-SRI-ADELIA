```python
import pandas as pd # pandas berfunsi untuk mengolah data dalam bentuk tabel (dataframe)
from scipy.stats.mstats import gmean, hmean # gmean untuk menghitung geometric mean, hmean untuk menghitung harmonic mean
import matplotlib.pyplot as plt # matplotlib untuk membuat visualisasi data, dalam hal ini histogram dan garis vertikal untuk mean

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel, pastikan file tersebut berada di direktori yang sama dengan script ini atau berikan path yang benar

a_mean = my_dataset.Zr.mean() # menghitung arithmetic mean dari kolom 'Zr' dalam dataset
g_mean = gmean(my_dataset['Zr']) # menghitung geometric mean dari kolom 'Zr' dalam dataset
h_mean = hmean(my_dataset['Zr']) # menghitung harmonic mean dari kolom 'Zr' dalam dataset

print ('-------') # mencetak garis pemisah untuk memperjelas output
print ('arithmetic mean') # mencetak judul untuk hasil arithmetic mean
print ("{0:.1f} [ppm]".format(a_mean)) # mencetak hasil arithmetic mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------')

print ('geometric mean') # mencetak judul untuk hasil geometric mean
print ("{0:.1f} [ppm]".format(g_mean)) # mencetak hasil geometric mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------')

print ('harmonic mean') # mencetak judul untuk hasil harmonic mean
print ("{0:.1f} [ppm]".format(h_mean)) # mencetak hasil harmonic mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------')

fig, ax = plt.subplots() # membuat figure dan axes untuk plot, fig adalah objek figure utama, ax adalah objek axes yang digunakan untuk membuat plot
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) # membuat histogram dari kolom 'Zr' dalam dataset, bins='auto' untuk menentukan jumlah bin secara otomatis, density=True untuk menormalkan histogram sehingga area di bawah histogram sama dengan 1, edgecolor='k' untuk memberikan warna hitam pada tepi bin, label untuk memberikan label pada legend, alpha untuk mengatur transparansi warna histogram
ax.axvline(a_mean, color='pink', label='Arithmetic mean', linewidth=2) # menambahkan garis vertikal pada posisi arithmetic mean, color untuk menentukan warna garis, label untuk memberikan label pada legend, linewidth untuk menentukan ketebalan garis
ax.axvline(g_mean, color='blue',  label='Geometric mean', linewidth=2) # menambahkan garis vertikal pada posisi geometric mean, color untuk menentukan warna garis, label untuk memberikan label pada legend, linewidth untuk menentukan ketebalan garis
ax.axvline(h_mean, color='yellow',  label='Harmonic mean', linewidth=2) # menambahkan garis vertikal pada posisi harmonic mean, color untuk menentukan warna garis, label untuk memberikan label pada legend, linewidth untuk menentukan ketebalan garis
ax.set_xlabel('Zr [ppm]') # memberikan label pada sumbu x, dalam hal ini 'Zr [ppm]' untuk menunjukkan bahwa data yang diukur adalah konsentrasi Zr dalam satuan ppm
ax.set_ylabel('Probability density') # memberikan label pada sumbu y, dalam hal ini 'Probability density' untuk menunjukkan bahwa nilai pada sumbu y adalah kerapatan probabilitas dari data yang diukur
ax.legend() # menampilkan legend pada plot, legend akan menampilkan label yang telah diberikan pada histogram dan garis vertikal untuk masing-masing mean

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd # pandas berfunsi untuk mengolah data dalam bentuk tabel (dataframe)
import numpy as np # numpy untuk melakukan operasi matematika pada array, dalam hal ini untuk menghitung histogram dan mencari modal value
import matplotlib.pyplot as plt # matplotlib untuk membuat visualisasi data, dalam hal ini histogram dan garis vertikal untuk modal value

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel, pastikan file tersebut berada di direktori yang sama dengan script ini atau berikan path yang benar

hist, bin_edges = np.histogram(my_dataset['Zr'], bins= 20, density=True) # menghitung histogram dari kolom 'Zr' dalam dataset, bins=20 untuk menentukan jumlah bin sebanyak 20, density=True untuk menormalkan histogram sehingga area di bawah histogram sama dengan 1, hist adalah array yang berisi nilai kerapatan probabilitas untuk masing-masing bin, bin_edges adalah array yang berisi batas-batas bin
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 # mencari modal value, hist.argmax() akan mengembalikan indeks dari bin yang memiliki nilai kerapatan probabilitas tertinggi, bin_edges[hist.argmax()] adalah batas bawah dari bin tersebut, bin_edges[hist.argmax()+1] adalah batas atas dari bin tersebut, dengan mengambil rata-rata dari kedua batas tersebut kita mendapatkan modal value yang merupakan nilai tengah dari bin dengan kerapatan probabilitas tertinggi

print ('modal value: {0:.0f} [ppm]'.format(modal_value)) # mencetak hasil modal value dengan format tanpa angka di belakang koma dan satuan ppm

fig, ax = plt.subplots() # membuat figure dan axes untuk plot, fig adalah objek figure utama, ax adalah objek axes yang digunakan untuk membuat plot
ax.hist(my_dataset.Zr, bins=20, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8) # membuat histogram dari kolom 'Zr' dalam dataset, bins=20 untuk menentukan jumlah bin sebanyak 20, density=True untuk menormalkan histogram sehingga area di bawah histogram sama dengan 1, edgecolor='k' untuk memberikan warna hitam pada tepi bin, label untuk memberikan label pada legend, alpha untuk mengatur transparansi warna histogram
ax.axvline(modal_value, color="red", label="Modal value", linewidth=2) # menambahkan garis vertikal pada posisi modal value, color untuk menentukan warna garis, label untuk memberikan label pada legend, linewidth untuk menentukan ketebalan garis
ax.set_xlabel('Zr [ppm]') # memberikan label pada sumbu x, dalam hal ini 'Zr [ppm]' untuk menunjukkan bahwa data yang diukur adalah konsentrasi Zr dalam satuan ppm
ax.set_ylabel('Probability density') # memberikan label pada sumbu y, dalam hal ini 'Probability density' untuk menunjukkan bahwa nilai pada sumbu y adalah kerapatan probabilitas dari data yang diukur
ax.legend() # menampilkan legend pada plot, legend akan menampilkan label yang telah diberikan pada histogram dan garis vertikal untuk masing-masing modal value

'''
Output: modal value: 277 [ppm]
'''
```

```python
import pandas as pd # pandas berfunsi untuk mengolah data dalam bentuk tabel (dataframe)
import numpy as np # numpy untuk melakukan operasi matematika pada array, dalam hal ini untuk menghitung kuartil dan interkuartil range
import matplotlib.pyplot as plt # matplotlib untuk membuat visualisasi data, dalam hal ini histogram dan area yang menunjukkan range antara nilai minimum dan maksimum

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # membaca data dari file excel, pastikan file tersebut berada di direktori yang sama dengan script ini atau berikan path yang benar

q1 = np.percentile(my_dataset.Zr, 25, method='midpoint') # menghitung kuartil pertama (Q1) dari kolom 'Zr' dalam dataset, kuartil pertama adalah nilai yang membagi data menjadi 25% bagian bawah dan 75% bagian atas, method='midpoint' digunakan untuk menentukan nilai kuartil ketika jumlah data genap, dengan mengambil rata-rata dari dua nilai tengah
q3 = np.percentile(my_dataset.Zr, 75, method='midpoint') # menghitung kuartil ketiga (Q3) dari kolom 'Zr' dalam dataset, kuartil ketiga adalah nilai yang membagi data menjadi 75% bagian bawah dan 25% bagian atas, method='midpoint' digunakan untuk menentukan nilai kuartil ketika jumlah data genap, dengan mengambil rata-rata dari dua nilai tengah 
  
iqr = q3 - q1 # Interquaritle range (IQR)  

print ('-------') # mencetak garis pemisah untuk memperjelas output
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) # mencetak hasil interquaritle range dengan format tanpa angka di belakang koma dan satuan ppm
print ('-------') # mencetak garis pemisah untuk memperjelas output

fig, ax = plt.subplots()
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist') # membuat histogram dari kolom 'Zr' dalam dataset, bins='auto' untuk menentukan jumlah bin secara otomatis, density=True untuk menormalkan histogram sehingga area di bawah histogram sama dengan 1, edgecolor='k' untuk memberikan warna hitam pada tepi bin, label untuk memberikan label pada legend 
ax.axvline(q1, color='purple', label='Q1', linewidth=2) # menambahkan garis vertikal pada posisi kuartil pertama (Q1), color untuk menentukan warna garis, label untuk memberikan label pada legend, linewidth untuk menentukan ketebalan garis
ax.axvline(q3, color='white', label='Q3', linewidth=2) # menambahkan garis vertikal pada posisi kuartil ketiga (Q3), color untuk menentukan warna garis, label untuk memberikan label pada legend, linewidth untuk menentukan ketebalan garis
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') # menambahkan area transparan yang menunjukkan range antara kuartil pertama (Q1) dan kuartil ketiga (Q3), alpha untuk mengatur transparansi warna, color untuk menentukan warna area, label untuk memberikan label pada legend
ax.set_xlabel('Zr [ppm]') # memberikan label pada sumbu x, dalam hal ini 'Zr [ppm]' untuk menunjukkan bahwa data yang diukur adalah konsentrasi Zr dalam satuan ppm
ax.set_ylabel('Probability density') # memberikan label pada sumbu y, dalam hal ini 'Probability density' untuk menunjukkan bahwa nilai pada sumbu y adalah kerapatan probabilitas dari data yang diukur
ax.legend() # menampilkan legend pada plot, legend akan menampilkan label yang telah diberikan pada histogram dan garis vertikal untuk masing-masing kuartil serta area untuk interquaritle range

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import numpy as np # numpy untuk melakukan operasi matematika pada array, dalam hal ini untuk menghitung skewness
from scipy.stats import skew # skew untuk menghitung skewness dari data, skewness adalah ukuran ketidaksimetrian distribusi data, jika skewness > 0 maka distribusi data miring ke kanan, jika skewness < 0 maka distribusi data miring ke kiri, jika skewness = 0 maka distribusi data simetris

a_mean = my_dataset.Zr.mean() # menghitung arithmetic mean dari kolom 'Zr' dalam dataset
median = my_dataset.Zr.median() # menghitung median dari kolom 'Zr' dalam dataset
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True) # menghitung histogram dari kolom 'Zr' dalam dataset
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 # mencari modal value
standard_deviation = my_dataset['Zr'].std() # menghitung standard deviation dari kolom 'Zr' dalam dataset

a1 = (a_mean - modal_value) / standard_deviation if standard_deviation != 0 else 0 # menghitung Pearson's first coefficient of skewness
a2 = 3 * (a_mean - median) / standard_deviation if standard_deviation != 0 else 0 # menghitung Pearson's second coefficient of skewness
g1 = skew(my_dataset['Zr']) # menghitung Fisher-Pearson's coefficient of skewness

print ('-------') # mencetak garis pemisah untuk memperjelas output
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1)) # mencetak hasil Pearson's first coefficient of skewness dengan format dua angka di belakang koma
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2)) # mencetak hasil Pearson's second coefficient of skewness dengan format dua angka di belakang koma
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1)) # mencetak hasil Fisher-Pearson's coefficient of skewness dengan format dua angka di belakang koma
print ('-------') # mencetak garis pemisah untuk memperjelas output

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
