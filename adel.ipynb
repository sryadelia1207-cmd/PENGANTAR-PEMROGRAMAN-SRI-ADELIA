```python
import numpy as np # numpy untuk operasi matematika dan array, digunakan untuk membuat fungsi integrasi dan menghitung nilai integral
from scipy import integrate # scipy untuk fungsi integrasi yang lebih canggih seperti trapezoidal dan Simpson, digunakan untuk menghitung nilai integral dengan metode yang lebih akurat

def integrate_rec(f, a, b, n):
    # Implementation of the rectanlge method 
    h = (b-a)/n # menghitung lebar setiap subinterval dengan membagi panjang interval [a, b] dengan jumlah subinterval n
    x = np.linspace(a, b, n+1) # membuat array x yang berisi titik-titik pembagi interval [a, b] menjadi n subinterval, dengan menggunakan fungsi linspace dari numpy yang menghasilkan n+1 titik mulai dari a hingga b
    i=0 # inisialisasi variabel i untuk iterasi
    area=0 # inisialisasi variabel area untuk menyimpan hasil penjumlahan luas persegi panjang
    while i<n: # loop untuk iterasi sebanyak n kali, dimana i akan bertambah dari 0 hingga n-1
        sup_rect =  f(x[i])*h # menghitung luas persegi panjang dengan menggunakan nilai fungsi f pada titik x[i] sebagai tinggi dan h sebagai lebar, hasilnya disimpan dalam variabel sup_rect
        area += sup_rect # menambahkan luas persegi panjang yang baru dihitung ke variabel area, sehingga area akan menyimpan total luas dari semua persegi panjang yang dibentuk selama iterasi
        i += 1  
    return area # mengembalikan nilai total luas yang dihitung sebagai hasil dari metode rectangle
 r'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5) # menghitung nilai integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode rectangle dan membagi interval menjadi 5 subinterval, hasilnya disimpan dalam variabel sup_5
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10) # menghitung nilai integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode rectangle dan membagi interval menjadi 10 subinterval, hasilnya disimpan dalam variabel sup_10
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100) # menghitung nilai integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode rectangle dan membagi interval menjadi 100 subinterval, hasilnya disimpan dalam variabel sup_100

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5)) # mencetak hasil nilai integral yang dihitung dengan metode rectangle untuk n=5, menggunakan format string untuk menampilkan nilai dengan 2 angka di belakang koma
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10)) # mencetak hasil nilai integral yang dihitung dengan metode rectangle untuk n=10, menggunakan format string untuk menampilkan nilai dengan 2 angka di belakang koma
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100)) # mencetak hasil nilai integral yang dihitung dengan metode rectangle untuk n=100, menggunakan format string untuk menampilkan nilai dengan 2 angka di belakang koma
     
'''
Output:
Using n=5, the rectangle method returns a value of 0.83
Using n=10, the rectangle method returns a value of 0.92
Using n=100, the rectangle method returns a value of 0.99
'''
 
```

```python
import numpy as np # numpy untuk operasi matematika dan array, digunakan untuk membuat fungsi integrasi dan menghitung nilai integral

def integrate_trap(f, a, b, n): # mendefinisikan fungsi integrate_trap yang menerima empat parameter: f (fungsi yang akan diintegrasikan), a (batas bawah integral), b (batas atas integral), dan n (jumlah subinterval untuk metode trapezoidal)
    # Implementation of the trapezoidal rule
    h = (b-a)/n # menghitung lebar setiap subinterval dengan membagi panjang interval [a, b] dengan jumlah subinterval n
    x = np.linspace(a, b, n+1) # membuat array x yang berisi titik-titik pembagi interval [a, b] menjadi n subinterval, dengan menggunakan fungsi linspace dari numpy yang menghasilkan n+1 titik mulai dari a hingga b
    i=1 # inisialisasi variabel i untuk iterasi, dimulai dari 1 karena titik pertama (x[0]) dan titik terakhir (x[n]) akan dihitung secara terpisah dalam rumus trapezoidal
    area = h*(f(x[0]) + f(x[n]))/2 # menghitung kontribusi dari titik pertama dan titik terakhir dalam rumus trapezoidal, dimana f(x[0]) adalah nilai fungsi pada titik pertama dan f(x[n]) adalah nilai fungsi pada titik terakhir, hasilnya dikalikan dengan h/2 untuk mendapatkan kontribusi awal area
    while i<n: # loop untuk iterasi sebanyak n-1 kali, dimana i akan bertambah dari 1 hingga n-1, karena titik pertama dan terakhir sudah dihitung sebelumnya
        sup_rect = f(x[i])*h # menghitung kontribusi dari titik-titik tengah (x[i]) dalam rumus trapezoidal, dimana f(x[i]) adalah nilai fungsi pada titik tengah dan h adalah lebar subinterval, hasilnya disimpan dalam variabel sup_rect
        area += sup_rect  # menambahkan kontribusi dari titik-titik tengah ke variabel area, sehingga area akan menyimpan total luas yang dihitung selama iterasi
        i += 1 # menambahkan 1 ke variabel i untuk melanjutkan iterasi ke titik tengah berikutnya
    return area # mengembalikan nilai total luas yang dihitung sebagai hasil dari metode trapezoidal

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5) # menghitung nilai integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode trapezoidal dan membagi interval menjadi 5 subinterval, hasilnya disimpan dalam variabel sup_5
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10) # menghitung nilai integral dari fungsi sinus pada interval [0, pi/2] dengan menggunakan metode trapezoidal dan membagi interval menjadi 10 subinterval, hasilnya disimpan dalam variabel sup_10

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5)) # mencetak hasil nilai integral yang dihitung dengan metode trapezoidal untuk n=5, menggunakan format string untuk menampilkan nilai dengan 2 angka di belakang koma
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10)) # mencetak hasil nilai integral yang dihitung dengan metode trapezoidal untuk n=10, menggunakan format string untuk menampilkan nilai dengan 2 angka di belakang koma
   
'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
'''
 
```

```python
import numpy as np # numpy untuk operasi matematika dan array, digunakan untuk membuat fungsi integrasi dan menghitung nilai integral
from scipy import integrate # scipy untuk fungsi integrasi yang lebih canggih seperti trapezoidal dan Simpson, digunakan untuk menghitung nilai integral dengan metode yang lebih akurat


x = np.linspace(0,9, 3) # 3 divisions %*\textcolor{codegreen}{$[x_0,x_1,x_2]$}*), n=2 
y = x**2 # f(x) = x^2

sup_trapz = integrate.trapezoid(y,x) # menghitung nilai integral dari fungsi y terhadap x menggunakan metode trapezoidal dengan fungsi trapezoid dari scipy.integrate, hasilnya disimpan dalam variabel sup_trapz
sup_simps = integrate.simpson(y,x) # menghitung nilai integral dari fungsi y terhadap x menggunakan metode Simpson dengan fungsi simpson dari scipy.integrate, hasilnya disimpan dalam variabel sup_simps


print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz)) # mencetak hasil nilai integral yang dihitung dengan metode trapezoidal untuk n=2, menggunakan format string untuk menampilkan nilai sebagai bilangan bulat tanpa desimal
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps)) # mencetak hasil nilai integral yang dihitung dengan metode Simpson untuk n=2, menggunakan format string untuk menampilkan nilai sebagai bilangan bulat tanpa desimal

'''
Output:
Using n=2, the trapezoidal rule returns a value of 273
Using n=2, the composite Simpson rule returns a value of 243
'''
```

```python
import numpy as np # numpy untuk operasi matematika dan array, digunakan untuk membuat fungsi integrasi dan menghitung nilai integral
from scipy import integrate # scipy untuk fungsi integrasi yang lebih canggih seperti trapezoidal dan Simpson, digunakan untuk menghitung nilai integral dengan metode yang lebih akurat

conturs_areas = np.array([194135, 136366, 79745, 38335, 18450, 9635, 3895]) # membuat array conturs_areas yang berisi nilai-nilai luas kontur pada titik-titik tertentu, yang akan digunakan untuk menghitung volume dengan metode integrasi
x = np.array([0,25,50,75,100,125,150]) # membuat array x yang berisi nilai-nilai jarak atau posisi yang sesuai dengan luas kontur yang diberikan dalam conturs_areas, yang akan digunakan sebagai variabel independen dalam perhitungan integrasi untuk mendapatkan volume

vol_traps = integrate.trapezoid(conturs_areas, x) # menghitung volume dengan menggunakan metode trapezoidal, dimana conturs_areas adalah nilai fungsi yang akan diintegrasikan dan x adalah variabel independen, hasilnya disimpan dalam variabel vol_traps
vol_simps = integrate.simpson(conturs_areas, x) # menghitung volume dengan menggunakan metode Simpson, dimana conturs_areas adalah nilai fungsi yang akan diintegrasikan dan x adalah variabel independen, hasilnya disimpan dalam variabel vol_simps

print('The trapezoidal rule returns a volume of {:.0f} cubic meters'.format(vol_traps)) # mencetak hasil volume yang dihitung dengan metode trapezoidal, menggunakan format string untuk menampilkan nilai sebagai bilangan bulat tanpa desimal
print('The composite Simpson rule returns a volume of {:.0f} cubic meters'.format(vol_simps)) # mencetak hasil volume yang dihitung dengan metode Simpson, menggunakan format string untuk menampilkan nilai sebagai bilangan bulat tanpa desimal

'''
Output:
The trapezoidal rule returns a volume of 9538650 cubic meters
The composite Simpson rule returns a volume of 9431367 cubic meters   
'''
```

```python
import numpy as np # numpy untuk operasi matematika dan array, digunakan untuk membuat fungsi integrasi dan menghitung nilai integral
from scipy.integrate import trapezoid # scipy untuk fungsi integrasi yang lebih canggih seperti trapezoidal, digunakan untuk menghitung nilai integral dengan metode yang lebih akurat
import matplotlib.pyplot as plt # matplotlib untuk visualisasi data, digunakan untuk membuat grafik yang menunjukkan hubungan antara jarak dari pusat bumi dan densitas pada interior bumi

r = np.linspace(1,6400,6400) # membuat array r yang berisi nilai-nilai jarak dari pusat bumi hingga permukaan bumi, dengan menggunakan fungsi linspace dari numpy untuk menghasilkan 6400 titik mulai dari 1 km hingga 6400 km, yang akan digunakan sebagai variabel independen dalam perhitungan dan visualisasi densitas pada interior bumi

def density(): # mendefinisikan fungsi density yang akan menghitung dan mengembalikan array densitas pada interior bumi berdasarkan jarak dari pusat bumi
    ro_inner_core = np.linspace(13100, 12800, 1220) # membuat array ro_inner_core yang berisi nilai-nilai densitas pada inti dalam bumi, dengan menggunakan fungsi linspace dari numpy untuk menghasilkan 1220 titik mulai dari 13100 kg/m^3 hingga 12800 kg/m^3, yang akan digunakan sebagai bagian dari perhitungan densitas pada interior bumi
    ro_outer_core = np.linspace(12200, 9900, 2259) # membuat array ro_outer_core yang berisi nilai-nilai densitas pada inti luar bumi, dengan menggunakan fungsi linspace dari numpy untuk menghasilkan 2259 titik mulai dari 12200 kg/m^3 hingga 9900 kg/m^3, yang akan digunakan sebagai bagian dari perhitungan densitas pada interior bumi
    ro_lower_mantle = np.linspace(5600,4400,2171) # membuat array ro_lower_mantle yang berisi nilai-nilai densitas pada mantel bawah bumi, dengan menggunakan fungsi linspace dari numpy untuk menghasilkan 2171 titik mulai dari 5600 kg/m^3 hingga 4400 kg/m^3, yang akan digunakan sebagai bagian dari perhitungan densitas pada interior bumi
    ro_upper_mantle = np.linspace(4130,3400,720) # membuat array ro_upper_mantle yang berisi nilai-nilai densitas pada mantel atas bumi, dengan menggunakan fungsi linspace dari numpy untuk menghasilkan 720 titik mulai dari 4130 kg/m^3 hingga 3400 kg/m^3, yang akan digunakan sebagai bagian dari perhitungan densitas pada interior bumi
    ro_crust = np.linspace(3100,2700,30) # membuat array ro_crust yang berisi nilai-nilai densitas pada kerak bumi, dengan menggunakan fungsi linspace dari numpy untuk menghasilkan 30 titik mulai dari 3100 kg/m^3 hingga 2700 kg/m^3, yang akan digunakan sebagai bagian dari perhitungan densitas pada interior bumi
   
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust)) # menggabungkan semua array densitas yang telah dibuat (ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust) menjadi satu array ro_final menggunakan fungsi concatenate dari numpy, sehingga ro_final akan berisi nilai-nilai densitas untuk seluruh interior bumi dari inti hingga kerak
    
    return ro_final # mengembalikan array ro_final sebagai hasil dari fungsi density, yang berisi nilai-nilai densitas pada interior bumi berdasarkan jarak dari pusat bumi

ro = density() # memanggil fungsi density untuk mendapatkan array ro yang berisi nilai-nilai densitas pada interior bumi, yang akan digunakan untuk visualisasi hubungan antara jarak dari pusat bumi dan densitas

fig, ax = plt.subplots() # membuat objek figure dan axes menggunakan fungsi subplots dari matplotlib.pyplot, yang akan digunakan untuk membuat grafik yang menunjukkan hubungan antara jarak dari pusat bumi (r) dan densitas pada interior bumi (ro)
ax.plot(r,ro, label="Densities on the Earth's Interior") # membuat grafik dengan menggunakan metode plot pada objek axes (ax), dimana r adalah variabel independen yang mewakili jarak dari pusat bumi dan ro adalah variabel dependen yang mewakili densitas pada interior bumi, serta memberikan label "Densities on the Earth's Interior" untuk legenda grafik
ax.set_ylabel(r"Density [Kg/m$^3$]") # memberikan label pada sumbu y dengan menggunakan metode set_ylabel pada objek axes (ax), dimana r"Density [Kg/m$^3$]" adalah string yang akan ditampilkan sebagai label, dengan menggunakan raw string (r"") untuk memastikan bahwa karakter khusus seperti $^3$ ditampilkan dengan benar dalam format LaTeX untuk menunjukkan satuan densitas dalam kilogram per meter kubik
ax.set_xlabel("Distance from the Earth center r [km]") # memberikan label pada sumbu x dengan menggunakan metode set_xlabel pada objek axes (ax), dimana "Distance from the Earth center r [km]" adalah string yang akan ditampilkan sebagai label, menunjukkan bahwa sumbu x mewakili jarak dari pusat bumi dalam satuan kilometer
ax.legend() # menampilkan legenda pada grafik dengan menggunakan metode legend pada objek axes (ax), yang akan menampilkan label yang telah diberikan sebelumnya untuk grafik, sehingga memudahkan pembaca untuk memahami apa yang diwakili oleh garis pada grafik
ax.grid() # menambahkan grid pada grafik dengan menggunakan metode grid pada objek axes (ax), yang akan membantu pembaca untuk lebih mudah membaca nilai-nilai pada grafik dengan adanya garis-garis grid sebagai referensi
```

```python
def gravity(r): # mendefinisikan fungsi gravity yang menerima parameter r (jarak dari pusat bumi) dan akan menghitung serta mengembalikan nilai percepatan gravitasi pada jarak tersebut berdasarkan distribusi densitas pada interior bumi
    
    g = np.zeros(len(r)) # membuat array g yang berisi nilai-nilai percepatan gravitasi yang akan dihitung, dengan menggunakan fungsi zeros dari numpy untuk menghasilkan array dengan panjang yang sama dengan array r, dimana setiap elemen awalnya diisi dengan nilai 0, dan akan diupdate dengan nilai percepatan gravitasi yang dihitung dalam loop selanjutnya
    Gr = 6.67408e-11 # gravitational constant in m^3 kg^-1 s^-2, menyimpan nilai konstanta gravitasi universal dalam variabel Gr, yang akan digunakan dalam perhitungan percepatan gravitasi
    r = r * 1000 # from Km to m # mengubah satuan jarak r dari kilometer ke meter dengan mengalikan setiap elemen dalam array r dengan 1000, karena perhitungan gravitasi memerlukan satuan meter untuk konsistensi dengan satuan lainnya seperti densitas dan konstanta gravitasi
    
    for i in range(1,len(r)): # loop untuk iterasi mulai dari 1 hingga panjang array r, dimana i akan bertambah dari 1 hingga len(r)-1, untuk menghitung percepatan gravitasi pada setiap jarak r[i] berdasarkan distribusi densitas dari pusat bumi hingga jarak tersebut
        
        r1 = r[0:i] # membuat array r1 yang berisi nilai-nilai jarak dari pusat bumi hingga jarak r[i], dengan menggunakan slicing pada array r untuk mengambil elemen dari indeks 0 hingga i-1, sehingga r1 akan berisi semua jarak yang lebih kecil dari r[i] yang akan digunakan untuk menghitung kontribusi gravitasi dari massa di dalam radius tersebut
        ro1  = ro[0:i] # membuat array ro1 yang berisi nilai-nilai densitas pada interior bumi yang sesuai dengan jarak dalam array r1, dengan menggunakan slicing pada array ro untuk mengambil elemen dari indeks 0 hingga i-1, sehingga ro1 akan berisi semua nilai densitas yang lebih kecil dari r[i] yang akan digunakan untuk menghitung kontribusi gravitasi dari massa di dalam radius tersebut
        r2 = r1[i-1] # menyimpan nilai jarak r[i] dalam variabel r2, yang akan digunakan sebagai jarak dari pusat bumi untuk menghitung percepatan gravitasi pada titik tersebut, dimana r2 adalah jarak yang sedang dihitung dalam iterasi saat ini
    
        y = ro1*r1**2 # menghitung nilai y yang merupakan hasil perkalian antara densitas ro1 dan kuadrat dari jarak r1, dimana y akan digunakan sebagai fungsi yang akan diintegrasikan untuk mendapatkan massa total di dalam radius r2, karena massa dapat dihitung dengan mengintegrasikan densitas dikalikan dengan volume elemen (yang proporsional dengan r^2 dalam koordinat bola)
        y_int = trapezoid(y,r1) # menghitung nilai integral dari fungsi y terhadap r1 menggunakan metode trapezoidal dengan fungsi trapezoid dari scipy.integrate, hasilnya disimpan dalam variabel y_int, yang mewakili massa total di dalam radius r2 berdasarkan distribusi densitas pada interior bumi
        
        g1 = ((4 * np.pi*Gr)/(r2**2)) * y_int # menghitung nilai percepatan gravitasi g1 pada jarak r2 dengan menggunakan rumus g = (4 * pi * G / r^2) * integral(densitas * r^2 dr), dimana Gr adalah konstanta gravitasi, r2 adalah jarak dari pusat bumi, dan y_int adalah hasil integral yang mewakili massa total di dalam radius tersebut, sehingga g1 akan memberikan nilai percepatan gravitasi pada jarak r2 berdasarkan distribusi densitas pada interior bumi
        g[i] = g1 # menyimpan nilai percepatan gravitasi g1 yang telah dihitung ke dalam array g pada indeks i, sehingga g akan berisi nilai percepatan gravitasi untuk setiap jarak r[i] yang dihitung dalam loop
           
    return g # mengembalikan array g yang berisi nilai-nilai percepatan gravitasi yang telah dihitung untuk setiap jarak r[i] berdasarkan distribusi densitas pada interior bumi, sebagai hasil dari fungsi gravity

g = gravity(r) # memanggil fungsi gravity dengan parameter r untuk menghitung array g yang berisi nilai-nilai percepatan gravitasi pada setiap jarak r[i] berdasarkan distribusi densitas pada interior bumi, hasilnya disimpan dalam variabel g

fig, ax = plt.subplots() # membuat objek figure dan axes menggunakan fungsi subplots dari matplotlib.pyplot, yang akan digunakan untuk membuat grafik yang menunjukkan hubungan antara jarak dari pusat bumi (r) dan percepatan gravitasi (g) pada interior bumi
ax.plot(r,g) # membuat grafik dengan menggunakan metode plot pada objek axes (ax), dimana r adalah variabel independen yang mewakili jarak dari pusat bumi dan g adalah variabel dependen yang mewakili percepatan gravitasi pada interior bumi, sehingga grafik akan menunjukkan bagaimana percepatan gravitasi berubah dengan jarak dari pusat bumi berdasarkan distribusi densitas yang telah digunakan dalam perhitungan
ax.grid() # menambahkan grid pada grafik dengan menggunakan metode grid pada objek axes (ax), yang akan membantu pembaca untuk lebih mudah membaca nilai-nilai pada grafik dengan adanya garis-garis grid sebagai referensi
ax.set_ylabel(r'g [m/s$^2]$') # memberikan label pada sumbu y dengan menggunakan metode set_ylabel pada objek axes (ax), dimana r'g [m/s$^2]$' adalah string yang akan ditampilkan sebagai label, dengan menggunakan raw string (r"") untuk memastikan bahwa karakter khusus seperti $^2$ ditampilkan dengan benar dalam format LaTeX untuk menunjukkan satuan percepatan gravitasi dalam meter per detik kuadrat
ax.set_xlabel('Distance from the Earth center r [km]') # memberikan label pada sumbu x dengan menggunakan metode set_xlabel pada objek axes (ax), dimana 'Distance from the Earth center r [km]' adalah string yang akan ditampilkan sebagai label, menunjukkan bahwa sumbu x mewakili jarak dari pusat bumi dalam satuan kilometer
```

```python
def pressure(r, ro, g): # mendefinisikan fungsi pressure yang menerima tiga parameter: r (array jarak dari pusat bumi), ro (array densitas pada interior bumi), dan g (array percepatan gravitasi pada interior bumi), dan akan menghitung serta mengembalikan array tekanan pada interior bumi berdasarkan distribusi densitas dan percepatan gravitasi
    
    p = np.zeros(len(r)) # membuat array p yang berisi nilai-nilai tekanan yang akan dihitung, dengan menggunakan fungsi zeros dari numpy untuk menghasilkan array dengan panjang yang sama dengan array r, dimana setiap elemen awalnya diisi dengan nilai 0, dan akan diupdate dengan nilai tekanan yang dihitung dalam loop selanjutnya
    r = r *1000 # from Km to m # mengubah satuan jarak r dari kilometer ke meter dengan mengalikan setiap elemen dalam array r dengan 1000, karena perhitungan tekanan memerlukan satuan meter untuk konsistensi dengan satuan lainnya seperti densitas dan percepatan gravitasi
    
    for i in range(0,len(r)): # loop untuk iterasi mulai dari 0 hingga panjang array r, dimana i akan bertambah dari 0 hingga len(r)-1, untuk menghitung tekanan pada setiap jarak r[i] berdasarkan distribusi densitas dan percepatan gravitasi dari pusat bumi hingga jarak tersebut  
        r1 = r[i:len(r)] # membuat array r1 yang berisi nilai-nilai jarak dari r[i] hingga jarak terakhir dalam array r, dengan menggunakan slicing pada array r untuk mengambil elemen dari indeks i hingga len(r)-1, sehingga r1 akan berisi semua jarak yang lebih besar atau sama dengan r[i] yang akan digunakan untuk menghitung kontribusi tekanan dari massa di luar radius tersebut
        ro1 = ro[i:len(r)] # membuat array ro1 yang berisi nilai-nilai densitas pada interior bumi yang sesuai dengan jarak dalam array r1, dengan menggunakan slicing pada array ro untuk mengambil elemen dari indeks i hingga len(r)-1, sehingga ro1 akan berisi semua nilai densitas yang lebih besar atau sama dengan r[i] yang akan digunakan untuk menghitung kontribusi tekanan dari massa di luar radius tersebut
        g1 = g[i:len(r)] # membuat array g1 yang berisi nilai-nilai percepatan gravitasi pada interior bumi yang sesuai dengan jarak dalam array r1, dengan menggunakan slicing pada array g untuk mengambil elemen dari indeks i hingga len(r)-1, sehingga g1 akan berisi semua nilai percepatan gravitasi yang lebih besar atau sama dengan r[i] yang akan digunakan untuk menghitung kontribusi tekanan dari massa di luar radius tersebut
        y = ro1*g1 # menghitung nilai y yang merupakan hasil perkalian antara densitas ro1 dan percepatan gravitasi g1, dimana y akan digunakan sebagai fungsi yang akan diintegrasikan untuk mendapatkan tekanan total pada jarak r[i], karena tekanan dapat dihitung dengan mengintegrasikan densitas dikalikan dengan percepatan gravitasi di sepanjang jarak dari pusat bumi hingga permukaan
        p1 = trapezoid(y,r1) # menghitung nilai integral dari fungsi y terhadap r1 menggunakan metode trapezoidal dengan fungsi trapezoid dari scipy.integrate, hasilnya disimpan dalam variabel p1, yang mewakili tekanan total pada jarak r[i] berdasarkan distribusi densitas dan percepatan gravitasi pada interior bumi
        p[i] = p1 # menyimpan nilai tekanan p1 yang telah dihitung ke dalam array p pada indeks i, sehingga p akan berisi nilai tekanan untuk setiap jarak r[i] yang dihitung dalam loop
    return p # mengembalikan array p yang berisi nilai-nilai tekanan yang telah dihitung untuk setiap jarak r[i] berdasarkan distribusi densitas dan percepatan gravitasi pada interior bumi, sebagai hasil dari fungsi pressure
    
p = pressure(r,ro,g)/1e9 # expressed in GPa # memanggil fungsi pressure dengan parameter r, ro, dan g untuk menghitung array p yang berisi nilai-nilai tekanan pada interior bumi berdasarkan distribusi densitas dan percepatan gravitasi, kemudian membagi hasilnya dengan 1e9 untuk mengubah satuan tekanan dari Pascal ke GigaPascal, hasilnya disimpan dalam variabel p
z = np.linspace(6400, 1, 6400) # from the Earth Surface to the Earth Center # membuat array z yang berisi nilai-nilai kedalaman dari permukaan bumi hingga pusat bumi, dengan menggunakan fungsi linspace dari numpy untuk menghasilkan 6400 titik mulai dari 6400 km hingga 1 km, yang akan digunakan sebagai variabel independen untuk visualisasi hubungan antara kedalaman dan tekanan pada interior bumi

fig, ax = plt.subplots() # membuat objek figure dan axes menggunakan fungsi subplots dari matplotlib.pyplot, yang akan digunakan untuk membuat grafik yang menunjukkan hubungan antara kedalaman (z) dan tekanan (p) pada interior bumi
ax.plot(z,p) # membuat grafik dengan menggunakan metode plot pada objek axes (ax), dimana z adalah variabel independen yang mewakili kedalaman dari permukaan bumi dan p adalah variabel dependen yang mewakili tekanan pada interior bumi, sehingga grafik akan menunjukkan bagaimana tekanan berubah dengan kedalaman berdasarkan
ax.grid() # menambahkan grid pada grafik dengan menggunakan metode grid pada objek axes (ax), yang akan membantu pembaca untuk lebih mudah membaca nilai-nilai pada grafik dengan adanya garis-garis grid sebagai referensi
ax.set_ylabel('P [GPa]') # memberikan label pada sumbu y dengan menggunakan metode set_ylabel pada objek axes (ax), dimana 'P [GPa]' adalah string yang akan ditampilkan sebagai label, menunjukkan bahwa sumbu y mewakili tekanan dalam satuan GigaPascal
ax.set_xlabel('Depth z from the Earth Surface [km]') # memberikan label pada sumbu x dengan menggunakan metode set_xlabel pada objek axes (ax), dimana 'Depth z from the Earth Surface [km]' adalah string yang akan ditampilkan sebagai label, menunjukkan bahwa sumbu x mewakili kedalaman dari permukaan bumi dalam satuan kilometer
```
